# Busqueda dicotomica

### Funci√≥n de prueba

```python
import numpy as np
import matplotlib.pyplot as plt

def quasiconvex_function(x):
  return ((x-3.6)**2) +2

#Graficar funci√≥n
# Generar valores de x
x_values = np.linspace(-2, 10, 20)

# Calcular los valores correspondientes de y utilizando la funci√≥n quasiconvex_function
y_values = [quasiconvex_function(x) for x in x_values]

# Graficar la funci√≥n
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Funci√≥n Cuasiconvexa')
plt.title('Gr√°fica de la Funci√≥n Cuasiconvexa')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
```

### Desarrollo del algoritmo

```python
#Selecci√≥n de intervalo [a, b]
a = -2
b = 10

# Epsilon ( Constante de distinguibilidad ) Debe ser peque√±a para distinguir entre los puntos
e = 0.1

#Precisi√≥n deseada
precision = 0.5

while (b-a) > precision:

    # Selecci√≥n de puntos de prueba
    p1 = ((a+b)/2)-e
    p2 = ((a+b)/2)+e
    print(f"Intervalo actual: [{a}, {b}]")
    print(f"Puntos de prueba: p1 = {p1}, p2 = {p2}")
    print("--------------------------------------------------")

    # Evaluaci√≥n de la funci√≥n en los puntos de prueba
    f_p1 = quasiconvex_function(p1)
    f_p2 = quasiconvex_function(p2)

    print(f"p1: {p1}, f(p1): {f_p1}")
    print(f"p2: {p2}, f(p2): {f_p2}")

    if f_p1 < f_p2:
        b = p2
    else:
        a = p1
    
    print(f"El m√≠nimo se encuentra en el intervalo: [{a}, {b}]")
    print("==================================================")

```

```python
# Graficar la funci√≥n original y marcar el resultado de la b√∫squeda dicot√≥mica
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Funci√≥n Cuasiconvexa', color='blue')

# Marcar el intervalo final donde se encuentra el m√≠nimo
plt.axvspan(a, b, alpha=0.3, color='red', label=f'Intervalo final [{a:.3f}, {b:.3f}]')

# Marcar el punto medio del intervalo final como aproximaci√≥n del m√≠nimo
x_min_approx = (a + b) / 2
y_min_approx = quasiconvex_function(x_min_approx)
plt.plot(x_min_approx, y_min_approx, 'ro', markersize=8, label=f'M√≠nimo aproximado ({x_min_approx:.3f}, {y_min_approx:.3f})')

plt.title('Resultado de la B√∫squeda Dicot√≥mica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
